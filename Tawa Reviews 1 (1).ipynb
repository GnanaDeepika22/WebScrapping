```python
url = "https://www.flipkart.com/master-superior-base-tawa-28-cm-diameter/product-reviews/itm40db66b45699a?pid=PTPFTGGBBGUEJZNU&lid=LSTPTPFTGGBBGUEJZNUKQ5Z4Z&marketplace=FLIPKART&page=57"
```

```python
import numpy as np



reviewer_name = []
reviewer_rating = []
review_title = []
review_text = []
place_of_review = []
date_of_review = []
upvotes = []
downvotes = []

pages = 497
for i in range(1,pages):
    url = "https://www.flipkart.com/master-superior-base-tawa-28-cm-diameter/product-reviews/itm40db66b45699a?pid=PTPFTGGBBGUEJZNU&lid=LSTPTPFTGGBBGUEJZNUKQ5Z4Z&marketplace=FLIPKART&page={}".format(i)
    response_headers = {'Content-Type':'text/html; charset=UTF-8',
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36',
                   'Accept-Encoding':'gzip, deflate, br'}
    response = requests.get(url,headers=response_headers)
    html_code=response.text
    soup = BeautifulSoup(html_code)
    
    containers = soup.find_all('div',attrs={'class':"col _2wzgFH K0kLPL"})
    
    for container in containers:
       
        
        #Reviewer rating
        rating = container.find('div', attrs={'class' : "_3LWZlK _1BLPMq"})
        if rating is None:
            reviewer_rating.append(np.NaN)
        else:
            reviewer_rating.append(rating.text)
        
        #Review title
        title = container.find('p',attrs={'class' : "_2-N8zT"})
        if title is None:
            review_title.append(np.NaN)
        else:
            review_title.append(title.text)
        
        #Review text
        text = container.find('div',attrs = {'class' : "t-ZTKy"})
        if text is None:
            review_text.append(np.NaN)
        else:
            review_text.append(text.text)
            
        #Place of review
        place = container.find('p', attrs = {'class' : "_2mcZGG"})
       
        if place is None:
            place_of_review.append(np.NaN)
        else:
            place_of_review.append(place.text)
        
        #Date of review and reviewer name
        #Filtering the customer name and date
        reviewer_date = container.find_all('p', attrs = {'class' : "_2sc7ZR"})
        
        #reviewer name
        reviewer = reviewer_date[0]
        if reviewer is None:
            reviewer_name.append(np.NaN)
        else:
            reviewer_name.append(reviewer.text)
        
        #date of review
        date = reviewer_date[1]
        if date is None:
            date_of_review.append(np.NaN)
        else:
            date_of_review.append(date.text)
            
        #Upvotes
        Upvote = container.find('span',attrs={'class' : "_3c3Px5"})
        if Upvote is None:
            upvotes.append(np.NaN)
        else:
            upvotes.append(Upvote.text)
            
        #Downvotes
        Downvote = container.find('span', attrs={'class' : "_3c3Px5"})
        if Downvote is None:
            downvotes.append(np.NaN)
        else:
            downvotes.append(Downvote.text)
    print('.',end="")
```

```python
print(len(reviewer_name))
print(len(reviewer_rating))
print(len(review_title))
print(len(review_text))
print(len(place_of_review))
print(len(date_of_review))
print(len(upvotes))
print(len(downvotes))
```

```python
import pandas as pd

df = pd.DataFrame({'reviewer_name':reviewer_name,
'reviewer_rating':reviewer_rating,
'review_title':review_title,
'review_text':review_text,
'place_of_review':place_of_review,
'date_of_review':date_of_review,
'upvotes':upvotes,
'downvotes':downvotes})
```
